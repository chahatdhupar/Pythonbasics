Question 1: What is hypothesis testing in statistics?

  Hypothesis testing is a statistical method used to make decisions about population parameters based on sample data.
  It involves:
  Formulating a null hypothesis (H0).
  Formulating an alternative hypothesis (H1).
  Calculating a test statistic.
  Comparing the statistic to a critical value to decide whether to reject H0.




Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

  Null Hypothesis (H0):
  A statement of no effect or no difference; it is what we assume to be true until evidence suggests otherwise.

  Alternative Hypothesis (H1):
  Represents what we want to prove; it suggests that there is an effect or a difference.

  Difference:
  H0 asserts the status quo, while H1 posits a new effect or relationship.




Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

  Significance Level (α):
  A threshold set before the test, typically 0.05 or 0.01, that defines the probability of rejecting the null hypothesis when it is true.

  Role:
  It helps determine the critical value against which the test statistic is compared.
  If the p-value is less than α, reject H0; otherwise, do not reject it.




Question 4: What are Type I and Type II errors? Give examples of each

  Type I Error (α):
  Rejecting a true null hypothesis.
  Example: Concluding a new drug is effective when it really is not.
  
  Type II Error (β):
  Failing to reject a false null hypothesis.
  Example: Concluding a new drug is not effective when it actually is.




Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.


  Z-test:
  Used when the population variance is known or the sample size is large (n > 30).
  Assumes a normal distribution of the sample.
  
  T-test:
  Used when the population variance is unknown and sample sizes are small (n ≤ 30).
  More accurate for estimating the mean when sample sizes are small.









Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.) Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5

# Generating binomial distribution
data = np.random.binomial(n, p, 1000)

# Plotting histogram
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50  # Hypothesized mean
alpha = 0.05

# Sample statistics
mean_sample = np.mean(sample_data)
std_sample = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-test
z_stat = (mean_sample - mu_0) / (std_sample / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean = {mean_sample:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

if p_value < alpha:
    print("Reject H0: Significant evidence mean ≠ 50")
else:
    print("Fail to reject H0: No significant evidence mean ≠ 50")
```

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. (Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
mu = 50
sigma = 5
sample_size = 1000

# Simulating data
data = np.random.normal(mu, sigma, sample_size)

# Calculating 95% Confidence Interval
conf_interval = stats.norm.interval(0.95, loc=np.mean(data), scale=np.std(data, ddof=1)/np.sqrt(sample_size))

# Plotting data
plt.hist(data, bins=30, alpha=0.7, color='green')
plt.axvline(conf_interval[0], color='red', linestyle='dashed', linewidth=1)
plt.axvline(conf_interval[1], color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid()
plt.show()

print(f"95% Confidence Interval: {conf_interval}")
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. (Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data = np.random.normal(100, 15, 200)

# Z-scores
mean, std = np.mean(data), np.std(data)
z_scores = (data - mean) / std

print("First 5 Z-scores:", z_scores[:5])

# Plot standardized data
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.title('Standardized Data (Z-scores)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.show()

# Interpretation:
# Z-scores tell how many std. deviations each value is from the mean.
# Z=0 means at mean, Z=+1 means 1 std. above mean, Z=-2 means 2 std. below mean.
```
