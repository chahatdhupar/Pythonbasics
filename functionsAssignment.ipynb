1. What is the difference between a function and a method in Python?

  ->Function: A function is a standalone block of code that performs a specific task. It is defined using the def keyword and can be called independently of any object. Example:
  # Function
  def add(a, b):
      return a + b
  # Calling the function
  result_function = add(3, 5)  # Output: 8

  Method: A method is a function that is associated with an object (an instance of a class). It is defined within a class and is called on an instance of that class. Example:
  # Method
  class Calculator:
      def add(self, a, b):
          return a + b
  # Creating an instance of the class
  calc = Calculator()
  # Calling the method
  result_method = calc.add(3, 5)  # Output: 8

2. Explain the concept of function arguments and parameters in Python.

  ->In Python, functions can accept inputs known as arguments, which allow you to pass data into the function for processing. The terms "parameters" and "arguments" are often used interchangeably, but they have distinct meanings in the context of functions.

  Parameters are the variables defined in the function signature (the function definition) that specify what kind of arguments the function can accept. They act as placeholders for the values that will be passed to the function when it is called.Example:

   def greet(name):  # 'name' is a parameter
      return f"Hello, {name}!"

  Arguments are the actual values or data you pass to a function when you call it. These values are assigned to the corresponding parameters defined in the function.Example:

  message = greet("Alice")  # "Alice" is an argument
  print(message)  # Output: Hello, Alice!


3. What are the different ways to define and call a function in Python?

  ->1. Basic Function Definition and Call:

  The most straightforward way to define and call a function is by using the def keyword.

  Definition:
  def greet(name):
  return f"Hello, {name}!"

  Call:
  message = greet("Alice")
  print(message) # Output: Hello, Alice!

  2. Function with Default Parameters
  You can define functions with default parameter values. If an argument is not provided for a parameter with a default value, the default is used.

  Definition:
  def greet(name="Guest"):
  return f"Hello, {name}!"
  
  Call:
  print(greet()) # Output: Hello, Guest!
  print(greet("Bob")) # Output: Hello, Bob!

  3. Function with Variable-Length Arguments
  You can define functions that accept a variable number of arguments using *args for non-keyword arguments and **kwargs for keyword arguments.

  Definition with *args:
  def sum_all(*args):
  return sum(args)

  Call:
  total = sum_all(1, 2, 3, 4) # Output: 10

  Definition with **kwargs:
  def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

  Call:
  print_info(name="Alice", age=25) # Output: name: Alice, age: 25

  4. Lambda Functions
  Lambda functions are small anonymous functions defined using the lambda keyword. They can take any number of arguments but can only have one expression.

  Definition:

  add = lambda x, y: x + y
  
  Call:
  result = add(3, 5) # Output: 8

  5. Nested Functions
  You can define functions inside other functions. The inner function can access variables from the outer function.

  Definition:
  def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function
  
  Call:

  add_five = outer_function(5)
  result = add_five(3) # Output: 8

5. What are iterators in Python and how do they differ from iterables?

  ->An iterator is an object that represents a stream of data. It is an object that implements the iterator protocol, which consists of two methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, and the __next__() method returns the next value from the stream. When there are no more items to return, __next__() raises a StopIteration exception.

  Iterators maintain their state, meaning they remember where they are in the iteration process. Once an iterator has been exhausted (i.e., all items have been iterated over), it cannot be reset or reused. Example:

  my_list = [1, 2, 3]
  my_iterator = iter(my_list)  # Create an iterator from the list
  print(next(my_iterator))  # Output: 1
  print(next(my_iterator))  # Output: 2
  print(next(my_iterator))  # Output: 3
  # print(next(my_iterator))  # Raises StopIteration

  On the other hand, an iterable is any Python object that can return its elements one at a time. This means that you can loop over the object using a for loop or use it in a context that requires iteration (like a list comprehension). Common examples of iterables include lists, tuples, dictionaries, sets, and strings.



6. Explain the concept of generators in Python and how they are defined.

  ->A generator is a function that produces a sequence of values using the yield statement instead of return. When a generator function is called, it does not execute the function body immediately. Instead, it returns a generator object that can be iterated over.
  Generators use lazy evaluation, meaning they generate values on-the-fly and only when requested. This makes them memory-efficient, especially when dealing with large datasets or infinite sequences, as they do not store all values in memory at once.Example:

  def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Yield the current count
        count += 1   # Increment the count

7. What are the advantages of using generators over regular functions?
  
  -> The advantages of using generators over regular functions include:

    1. Memory Efficiency: Generators use less memory by yielding values one at a time.
    2. Improved Performance: They can start producing output immediately, reducing overhead.
    3. Simplified Code: Generators provide a cleaner and more readable syntax for creating iterators.
    4. Handling Infinite Sequences: Generators can produce infinite sequences without consuming memory.
    5. State Retention: They automatically maintain their state between calls, simplifying complex iteration logic.

    def square_numbers(numbers):
    for n in numbers:
        yield n * n

    def filter_even(numbers):
        for n in numbers:
            if n % 2 == 0:
                yield n

    # Using the generators
    numbers = range(10)
    squared = square_numbers(numbers)
    even_squared = filter_even(squared)

    for num in even_squared:
        print(num)  # Output: 0, 4, 16, 36, 64

9. Explain the purpose and usage of the `map()` function in Python.

  ->1. Transformation: The main purpose of map() is to transform each element of an iterable by applying a function to it. This is particularly useful when you want to perform the same operation on every item in a collection.

  2. Conciseness: Using map() can lead to more concise and readable code compared to using a loop to achieve the same result.

  3. Efficiency: map() is generally more efficient than using a loop because it is implemented in C and can take advantage of optimizations.

  # Define a function to square a number
  def square(x):
      return x ** 2

  # Create a list of numbers
  numbers = [1, 2, 3, 4, 5]

  # Use map() to apply the square function to each element
  squared_numbers = map(square, numbers)

  # Convert the map object to a list and print the result
  print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  ->1. map()
  Purpose: The map() function is used to apply a specified function to each item in an iterable (like a list or tuple) and return a new iterable (specifically, a map object) containing the results.

  Syntax:
  map(function, iterable, ...)

  Returns: A map object (which is an iterator) that can be converted to a list or other iterable types.

  Use Case: When you want to transform each element of an iterable using a function.

  Example:

  numbers = [1, 2, 3, 4]
  squared = map(lambda x: x ** 2, numbers)  # Output: [1, 4, 9, 16]
  print(list(squared))

  2. filter()
  Purpose: The filter() function is used to filter elements from an iterable based on a specified function that returns True or False. It returns only those elements for which the function returns True.

  Syntax:
  filter(function, iterable)

  Returns: A filter object (which is an iterator) that can be converted to a list or other iterable types.

  Use Case: When you want to select a subset of elements from an iterable based on a condition.

  Example:
  numbers = [1, 2, 3, 4, 5]
  even_numbers = filter(lambda x: x % 2 == 0, numbers)  # Output: [2, 4]
  print(list(even_numbers))

  3. reduce()
  Purpose: The reduce() function is used to apply a specified function cumulatively to the items of an iterable, reducing the iterable to a single value. It is not a built-in function in Python 3 and needs to be imported from the functools module.

  Syntax:
  from functools import reduce
  reduce(function, iterable[, initializer])
  
  Returns: A single value that is the result of applying the function cumulatively to the items of the iterable.

  Use Case: When you want to combine all elements of an iterable into a single value, such as summing numbers or concatenating strings.

  Example:
  from functools import reduce
  numbers = [1, 2, 3, 4]
  sum_result = reduce(lambda x, y: x + y, numbers)  # Output: 10
  print(sum_result)

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def check_prime(n):
  if n <= 1:
        return False
  for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
  return True

for number in range(1, 201):
    if check_prime(number):
        print(number, "is a prime number.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        fib = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib

fib = FibonacciIterator(10)
for number in fib:
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i


for val in powers_of_two(4):
    print(val)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line('example.txt'):
    print(line)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32


celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Temperatures in Fahrenheit:", fahrenheit_temps)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):
    def is_vowel(char):
        return char.lower() in 'aeiou'

    non_vowels = filter(lambda char: not is_vowel(char), input_string)

    return ''.join(non_vowels)

input_str = "Hello, World!"
result = remove_vowels(input_str)
print("Original String:", input_str)
print("Without Vowels:", result)
```
