1. What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

Definition: Deep Learning is a subset of machine learning (ML) that uses artificial neural networks with multiple layers (hence "deep") to model and learn complex patterns from data. It mimics the human brain's structure, enabling tasks like image recognition, natural language processing, and prediction by automatically extracting features from raw input.

Key Components: Relies on layers of interconnected nodes (neurons), where each layer processes data and passes it to the next. Training involves adjusting weights via algorithms like backpropagation to minimize errors.

Evolution of Deep Learning
Early Foundations (1940s-1950s): Inspired by neuroscience; Frank Rosenblatt invented the perceptron (a single-layer neural network) in 1957, capable of simple binary classification.
Stagnation and Revival (1960s-1980s): Initial hype faded due to limitations (e.g., XOR problem). Geoffrey Hinton and others revived it in the 1980s with backpropagation, allowing multi-layer networks, but computational limits persisted.
Breakthrough Era (2000s-2010s): Big data, GPUs, and algorithms like convolutional neural networks (CNNs) for images (e.g., AlexNet in 2012) and recurrent networks (RNNs) for sequences enabled breakthroughs. Frameworks like TensorFlow (2015) and PyTorch made it accessible.
Modern Era (2020s): Integration with transformers (e.g., GPT models) for language tasks; focus on efficiency, ethics, and applications in AI like self-driving cars and healthcare.

Differences from Traditional Machine Learning
Model Complexity: Traditional ML uses shallow models (e.g., linear regression, decision trees, SVMs) with manual feature engineering. Deep Learning automates feature extraction through deep networks, handling raw, unstructured data (e.g., images, text) without preprocessing.
Data and Compute Requirements: Traditional ML works with smaller datasets and less compute. Deep Learning needs large datasets (millions of examples) and high-performance hardware (GPUs/TPUs) for training, making it more resource-intensive.
Performance and Applications: Traditional ML excels at structured data and interpretable models. Deep Learning outperforms on complex, non-linear tasks (e.g., computer vision, speech recognition) but is less interpretable ("black box") and prone to overfitting without regularization.


2. Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

Definition: A Perceptron is the simplest form of an artificial neural network, introduced by Frank Rosenblatt in 1957. It acts as a binary classifier, mimicking a single neuron in the brain.

Components:
Inputs (x1, x2, ..., xn): Features or data points fed into the model.
Weights (w1, w2, ..., wn): Adjustable parameters that determine the importance of each input.
Bias (b): A constant added to the weighted sum to shift the decision boundary.

Activation Function: Typically a step function (e.g., Heaviside function) that outputs 1 if the input is above a threshold, otherwise 0.
Output (y): The predicted class (0 or 1).
Structure: Single-layer architecture with no hidden layers—inputs connect directly to the output via weights and bias.

Example: For classifying points above/below a line, the Perceptron adjusts weights to find the optimal linear boundary.

Limitations of a Perceptron
Linear Separability: Can only classify linearly separable data (e.g., AND gate works, but XOR does not, as XOR requires a non-linear boundary).
Single Layer: Lacks hidden layers, so it can't model complex, non-linear relationships or hierarchical features.
No Multi-Class Support: Designed for binary classification; extensions exist but are limited.
Sensitivity to Noise: Prone to overfitting on noisy data and may not converge if data isn't separable.
No Probabilistic Output: Outputs are binary (0/1), not probabilities, limiting uncertainty handling.
Historical Impact: These limitations led to the "AI winter" in the 1960s, paving the way for multi-layer networks (e.g., backpropagation) in modern deep learning.

3. Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

Purpose of Activation Functions in Neural Networks

Definition: Activation functions are mathematical operations applied to the output of each neuron in a neural network. They introduce non-linearity, enabling the network to learn complex patterns and relationships beyond simple linear transformations.

Key Roles:
Non-Linearity: Without them, the network would behave like a linear regression model, unable to capture non-linear data distributions (e.g., XOR problems).

Output Control: They map inputs to a specific range (e.g., 0-1 for probabilities), aiding in tasks like classification.
Gradient Flow: Influence backpropagation by affecting how gradients are computed, impacting learning stability and speed.

Feature Extraction: Allow hierarchical learning in deep networks by activating or suppressing signals.
Comparison of Sigmoid, ReLU, and Tanh Functions
Sigmoid (Logistic Function):


Output Range: 0 to 1
Pros: Smooth gradient; useful for binary classification (e.g., output as probability); interpretable for decision boundaries.
Cons: Vanishing gradient problem (gradients near 0 or 1 are small, slowing learning in deep networks); not zero-centered; computationally expensive.
Use Cases: Output layers in binary classification; older networks like perceptrons.

ReLU (Rectified Linear Unit):


Output Range: 0 to ∞
Pros: Simple and fast computation; mitigates vanishing gradients; promotes sparse activation (neurons "die" for negative inputs, aiding efficiency); widely used in deep networks.
Cons: Dying ReLU problem (neurons stuck at 0 if inputs are always negative); not zero-centered; unbounded output can cause exploding gradients.
Use Cases: Hidden layers in CNNs and deep networks (e.g., image recognition); default for many modern architectures.
Tanh (Hyperbolic Tangent):



Output Range: -1 to 1


Pros: Zero-centered (better than Sigmoid for symmetry); stronger gradients than Sigmoid; handles negative values well.
Cons: Still suffers from vanishing gradients in deep networks; more computationally intensive than ReLU.
Use Cases: Hidden layers in RNNs (e.g., sequence modeling); situations needing symmetric outputs around zero.


4. What is the difference between Loss function and Cost function in neural
networks? Provide examples.

In neural networks, both the loss function and the cost function are measures of how well the model is performing, but they differ in scope and purpose.

A loss function quantifies the error for a single training example. It measures the difference between the model’s predicted output and the actual target value for one sample. In other words, it tells us how wrong the model’s prediction is for that specific instance.

The cost function, on the other hand, represents the average of all individual losses across the entire training dataset. It provides a single scalar value that reflects the overall performance of the model. This is the function that optimization algorithms such as gradient descent aim to minimize during training.

5.  What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.


Role of Optimizers in Neural Networks:
Optimizers play a crucial role in training neural networks by adjusting the model’s weights and biases to minimize the cost function. During training, the optimizer updates parameters based on the gradients (partial derivatives) of the cost function with respect to each parameter. Its goal is to find the optimal set of parameters that results in the lowest possible error. In short, optimizers control how fast and effectively a neural network learns.

1. Gradient Descent

Gradient Descent is the most basic optimization algorithm. It updates the parameters in the opposite direction of the gradient of the cost function.

Advantages: Simple and effective for small datasets.
Disadvantages: Can be slow, may get stuck in local minima, and performance depends heavily on the learning rate.

2. RMSprop (Root Mean Square Propagation)

RMSprop improves on standard Gradient Descent by adapting the learning rate for each parameter. It maintains a moving average of the squared gradients to normalize updates.

Advantages: Works well for non-stationary or noisy problems (like recurrent networks).
Disadvantages: May still oscillate near the minima.

3. Adam (Adaptive Moment Estimation)

Adam combines the benefits of Momentum and RMSprop. It maintains moving averages of both gradients and their squares, providing adaptive learning rates and faster convergence.

Advantages: Fast convergence, efficient on large datasets, and less sensitive to hyperparameters.
Disadvantages: Can sometimes lead to overfitting or suboptimal minima.




6. Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.

```python
import numpy as np

# Define the input data and labels for the AND gate
# Inputs: [x1, x2], Outputs: x1 AND x2
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 0, 0, 1])  # Expected outputs

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand(1)
learning_rate = 0.1
epochs = 20

# Activation function: Step function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        # Compute weighted sum
        linear_output = np.dot(X[i], weights) + bias
        # Apply activation
        predicted = step_function(linear_output)
        # Calculate error
        error = y[i] - predicted
        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error
    print(f"Epoch {epoch+1}/{epochs}, Weights: {weights}, Bias: {bias}")

# Testing the trained perceptron
print("\nFinal Results:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    predicted = step_function(linear_output)
    print(f"Input: {X[i]} → Predicted Output: {predicted[0]} → Expected: {y[i]}")
```

7.  Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 200)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot activation functions
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.plot(x, y_relu, label='ReLU', color='red')
plt.plot(x, y_tanh, label='Tanh', color='green')

plt.title('Activation Functions: Sigmoid, ReLU, and Tanh')
plt.xlabel('Input (x)')
plt.ylabel('Output (f(x))')
plt.grid(True)
plt.legend()
plt.show()
```

8. : Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.


```python
# Import necessary libraries
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the input data
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode the target labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Build the neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),          # Flatten 2D images into 1D
    Dense(128, activation='relu'),          # Hidden layer with ReLU
    Dense(64, activation='relu'),           # Another hidden layer
    Dense(10, activation='softmax')         # Output layer (10 classes)
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=128, verbose=1)

# Print training accuracy
train_acc = history.history['accuracy'][-1]
print(f"\nTraining Accuracy: {train_acc * 100:.2f}%")
```

 9. Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.


```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load the Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalize the data (scale pixel values between 0 and 1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode target labels
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Build a simple neural network model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=128, validation_split=0.2, verbose=1)

# Plot training and validation loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', color='orange')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy', color='green')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.show()

```

10. You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.

Explain your real-time data science workflow:

● How would you design a deep learning model (perceptron or multilayer NN)?

● Which activation function and loss function would you use, and why?

● How would you train and evaluate the model, considering class imbalance?

● Which optimizer would be suitable, and how would you prevent overfitting?


1. Model Design

Use a Multilayer Perceptron (MLP) since data is structured (tabular).

Architecture Example:

Input layer: number of neurons = number of features.

Hidden layers: 2–3 layers with 64–128 neurons each.

Output layer: 1 neuron (binary classification).

Flow:
Input → Dense → ReLU → Dense → ReLU → Dropout → Dense (Sigmoid)

2. Activation and Loss Functions

Hidden layers: ReLU → avoids vanishing gradient and speeds convergence.

Output layer: Sigmoid → outputs probability (0–1) for binary fraud detection.

Loss function:

Binary Cross-Entropy (BCE) — measures difference between predicted and actual probability.

If imbalance is severe → use Weighted BCE or Focal Loss to give more weight to minority (fraud) class.

3. Training and Evaluation (Handling Class Imbalance)

Techniques:

Resampling: Oversample minority class (SMOTE) or undersample majority.

Class weights: Pass class_weight parameter to the training function.

Evaluation metrics: Accuracy is misleading — use
Precision, Recall, F1-score, and especially AUC-ROC.

Train/Validation Split: Use stratified splitting to maintain class proportions.

4. Optimizer and Overfitting Prevention

Optimizer:

Use Adam — adaptive learning rates, efficient for noisy, sparse gradients.

Prevent Overfitting:

Dropout layers (e.g., 0.3–0.5).

Early stopping on validation loss.

L2 regularization (weight decay).

Cross-validation for robust performance.
