6. What are the differences between NumPy arrays and Python lists?

  ->1. Data Type Homogeneity
  NumPy Arrays: All elements in a NumPy array must be of the same data type (homogeneous). This allows for optimized performance and memory usage.
  Python Lists: Python lists can contain elements of different data types (heterogeneous), allowing for greater flexibility in the types of data stored.
  2. Performance and Efficiency
  NumPy Arrays: NumPy arrays are more memory-efficient and faster for numerical computations due to their fixed data type and contiguous memory allocation. They are optimized for performance, especially for large datasets and mathematical operations.
  Python Lists: Python lists are less efficient for numerical operations because they store references to objects rather than the objects themselves, leading to higher memory overhead and slower performance for large datasets.
  3. Functionality and Operations
  NumPy Arrays: NumPy provides a wide range of mathematical functions and operations that can be applied directly to arrays, including element-wise operations, broadcasting, and linear algebra functions. This makes NumPy arrays particularly powerful for scientific computing and data analysis.
  Python Lists: While Python lists support basic operations (e.g., appending, slicing), they lack the advanced mathematical capabilities of NumPy arrays. Operations on lists often require explicit loops or list comprehensions.
  4. Multidimensional Support
  NumPy Arrays: NumPy natively supports multidimensional arrays (e.g., 2D matrices, 3D tensors) through its ndarray object, making it easy to work with complex data structures.
  Python Lists: While Python lists can be nested to create multidimensional structures (e.g., lists of lists), they do not provide built-in support for multidimensional operations, making them less convenient for such tasks.
  5. Indexing and Slicing
  NumPy Arrays: NumPy arrays support advanced indexing and slicing techniques, including boolean indexing and fancy indexing, which allow for more complex data selection and manipulation.
  Python Lists: Python lists support basic indexing and slicing, but they do not offer the same level of flexibility and functionality as NumPy arrays.
  6. Memory Consumption
  NumPy Arrays: NumPy arrays are more memory-efficient due to their fixed data type and contiguous memory layout, which reduces overhead.
  Python Lists: Python lists consume more memory because they store references to objects, leading to additional overhead for each element.
  7. Use Cases
  NumPy Arrays: Ideal for numerical computations, scientific computing, data analysis, and machine learning tasks where performance and efficiency are critical.
  Python Lists: Suitable for general-purpose programming, where flexibility in data types and structures is more important than performance.

18. What are some advanced features of NumPy?

->NumPy is a powerful library for numerical computing in Python, and it offers several advanced features that enhance its functionality and performance. Here are some of the advanced features of NumPy:

Broadcasting:

Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes. It automatically expands the smaller array across the larger array to make their shapes compatible for element-wise operations.
Universal Functions (ufuncs):

NumPy provides a set of vectorized functions, known as universal functions, that operate element-wise on arrays. These functions are optimized for performance and can handle arrays of different shapes and sizes efficiently.
Advanced Indexing:

NumPy supports advanced indexing techniques, such as boolean indexing and fancy indexing. This allows for more flexible and powerful ways to access and manipulate array elements based on conditions or specific indices.
Structured Arrays:

Structured arrays allow you to create arrays with complex data types, similar to records in a database. Each element can contain multiple fields of different data types, making it easier to work with heterogeneous data.
Masked Arrays:

Masked arrays allow you to handle missing or invalid data by masking certain elements. This feature is useful for performing calculations while ignoring specific values.
Linear Algebra Functions:

NumPy includes a comprehensive set of linear algebra functions, such as matrix multiplication, eigenvalue decomposition, and singular value decomposition (SVD). These functions are essential for scientific computing and data analysis.
Fourier Transforms:

NumPy provides functions for computing discrete Fourier transforms (DFT) and their inverses, which are useful for signal processing and frequency analysis.
Random Number Generation:

NumPy includes a robust random number generation module that allows for the generation of random samples from various probability distributions. This is useful for simulations and statistical modeling.
Performance Optimization:

NumPy is implemented in C and optimized for performance. It uses contiguous memory allocation and efficient algorithms to perform operations quickly, making it suitable for large datasets.
Interoperability with Other Libraries:

NumPy arrays are the foundation for many other scientific computing libraries in Python, such as SciPy, Pandas, and Matplotlib. This interoperability allows for seamless integration and data sharing between libraries.
Memory Layout Control:

NumPy allows you to control the memory layout of arrays (C-contiguous or Fortran-contiguous), which can be important for performance optimization in certain applications.
N-dimensional Arrays:

NumPy supports n-dimensional arrays (ndarrays), enabling the representation and manipulation of multi-dimensional data structures, which is essential for various scientific and engineering applications.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

->Contiguous Memory Allocation: NumPy arrays are stored in contiguous blocks of memory, which allows for efficient access and manipulation of data. This contiguous memory layout enables faster iteration and slicing operations compared to Python lists, which can be composed of elements scattered throughout memory.

Homogeneous Data Types: NumPy arrays are homogeneous, meaning all elements are of the same data type. This allows NumPy to optimize memory usage and perform operations more efficiently. In contrast, Python lists can contain elements of different types, which adds overhead for type checking and can slow down operations.

Vectorized Operations: NumPy is designed for numerical computations and supports vectorized operations, which means that operations can be applied to entire arrays at once without the need for explicit loops. This leads to significant performance improvements, especially for large datasets.

Lower Overhead: NumPy is implemented in C and optimized for performance. The slicing operations in NumPy are executed at a lower level, reducing the overhead associated with Python's higher-level list operations. This results in faster execution times for slicing and other array manipulations.

Advanced Slicing Capabilities: NumPy provides advanced slicing capabilities, such as slicing with strides and multi-dimensional slicing, which are implemented efficiently. This allows for more complex operations to be performed quickly compared to the more limited slicing capabilities of Python lists.

Broadcasting: NumPy supports broadcasting, which allows for operations on arrays of different shapes without the need for explicit replication of data. This feature can lead to more efficient memory usage and faster computations.

4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Experience': [2, 5, 7, 10, 12]
}
df = pd.DataFrame(data)

correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

8. Create a line plot with multiple lines using MatplotlibA

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Generate sample data
x = np.linspace(0, 10, 100)  # X-values (0 to 10, 100 points)
y1 = np.sin(x)               # Line 1: Sine wave
y2 = np.cos(x)               # Line 2: Cosine wave
y3 = np.sin(x + 1.5)         # Line 3: Phase-shifted sine wave

# Step 2: Create the figure and axes
plt.figure(figsize=(10, 6))  # Set figure size (width, height)

# Step 3: Plot multiple lines with custom styles
plt.plot(x, y1,
         label='sin(x)',
         color='royalblue',
         linestyle='-',
         linewidth=2)

plt.plot(x, y2,
         label='cos(x)',
         color='crimson',
         linestyle='--',
         linewidth=2)

plt.plot(x, y3,
         label='sin(x + 1.5)',
         color='forestgreen',
         linestyle=':',
         linewidth=2)

# Step 4: Customize the plot
plt.title('Trigonometric Functions', fontsize=14, pad=20)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid lines
plt.legend(fontsize=10)                   # Show legend with labels

# Optional: Customize axes limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Step 5: Show or save the plot
plt.tight_layout()  # Adjust spacing
plt.savefig('multi_line_plot.png', dpi=300)  # Save high-res version
plt.show()  # Display the plot
```

11. Perform matrix multiplication using NumPy

```python
import numpy as np

# Step 1: Create two matrices (2D arrays)
matrix_a = np.array([[1, 2, 3],
                     [4, 5, 6]])

matrix_b = np.array([[7, 8],
                     [9, 10],
                     [11, 12]])

# Step 2: Perform matrix multiplication
# Method 1: Using np.dot()
result_dot = np.dot(matrix_a, matrix_b)

# Method 2: Using the @ operator
result_at = matrix_a @ matrix_b

# Step 3: Display the results
print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)
print("\nResult of Matrix Multiplication (using np.dot):")
print(result_dot)
print("\nResult of Matrix Multiplication (using @ operator):")
print(result_at)
```

13. Create a 3D scatter plot using Plotly

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Step 1: Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Z': np.random.randn(100),
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Size': np.random.uniform(10, 50, 100)  # For marker size
})

# Step 2: Create the 3D scatter plot
fig = px.scatter_3d(
    data,
    x='X', y='Y', z='Z',                 # Axis columns
    color='Category',                    # Color by category
    size='Size',                         # Marker size
    opacity=0.8,                         # Transparency
    title='3D Scatter Plot with Plotly', # Plot title
    template='plotly_dark',              # Theme (optional)
    height=700                           # Figure height
)

# Step 3: Customize layout
fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera_eye=dict(x=1.5, y=1.5, z=0.1)  # Adjust camera view
    ),
    margin=dict(l=0, r=0, b=0, t=40)         # Margins
)

# Step 4: Show the plot
fig.show()
```
