1.	What is a random variable in probability theory?

  A random variable is a variable whose possible values are numerical outcomes of a random phenomenon. It can be thought of as a function that maps outcomes of a random process to numbers.
  Example: If you roll a six-sided die, the outcome can be represented as a random variable $X$ where:
  
  X = 1 if the die shows 1,
  X = 2 if the die shows 2,
  ...
  X = 6 if the die shows 6.

2.	What are the types of random variables?

  Discrete Random Variable: Takes on a countable number of values. For example, the number of heads in 10 coin flips.
  Example: Let (Y) be the number of heads in 10 flips of a fair coin. (Y) can take values from 0 to 10.
  Continuous Random Variable: Takes on an infinite number of values within a range. For example, the height of students in a classroom.
  Example: Let (Z) be the height of students in a class. (Z) can take any value within a certain range, say from 150 cm to 200 cm.

3.	 Explain the difference between discrete and continuous distributions.

  Discrete Distribution: The probability distribution of a discrete random variable. The probabilities of each possible value sum to 1.

  Example: The binomial distribution is a discrete distribution. If you flip a coin 10 times, the probability of getting exactly 5 heads can be calculated using the binomial formula.

  Continuous Distribution: The probability distribution of a continuous random variable. The probability of the variable falling within a certain range is given by the area under the curve of the probability density function (PDF).

  Example: The normal distribution is a continuous distribution. The probability of a student scoring between 70 and 80 on a test can be found by calculating the area under the normal curve between those two scores.

4.	What is a binomial distribution, and how is it used in probability?

  A binomial distribution is a discrete probability distribution that models the probability of achieving a specific number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure). It's used to calculate probabilities in scenarios with repeated, independent trials, like coin flips or quality control checks.
  The binomial distribution helps determine the probability of getting exactly k successes in n trials, given the probability of success (p) in a single trial. This is calculated using the binomial probability formula:

  P(X=k) = (nCk) * p^k * (1-p)^(n-k), where nCk represents the number of combinations of choosing k successes from n trials.

5.	What is the standard normal distribution, and why is it important?

  The standard normal distribution is a normal probability distribution with a mean of 0 and a standard deviation of 1. It's a crucial tool because any normal distribution can be transformed into a standard normal distribution using z-scores, allowing for easier probability calculations and comparisons across different datasets.

  •	Simplifies Probability Calculations:
  By converting normal distributions to the standard normal distribution, we can use standard normal distribution tables (Z-tables) to easily find probabilities associated with specific z-scores.
  •	Enables Comparisons:
  It allows us to compare data from different normal distributions that may have varying means and standard deviations.
  •	Standardization:
  Converting data to z-scores standardizes the data, making it easier to analyze and compare, even when the original data has different units or scales.
  •	Foundation for Statistical Inference:
  The standard normal distribution is fundamental to many statistical concepts and techniques, including hypothesis testing and confidence intervals.
  •	Central Limit Theorem:
  The standard normal distribution is closely related to the Central Limit Theorem, which states that the distribution of sample means will approach a normal distribution, regardless of the original distribution, as the sample size increases. This makes the standard normal distribution a powerful tool for analyzing a wide range of data.


6.	 What is the Central Limit Theorem (CLT), and why is it critical in statistics?

  The Central Limit Theorem (CLT) states that the distribution of sample means will approximate a normal distribution, regardless of the original population's distribution, as long as the sample size is sufficiently large. This is critical in statistics because it allows us to make inferences about populations using sample data, even when the population distribution is unknown.
  
  Example: If you take multiple samples of size 30 from a population with any distribution (e.g., uniform, skewed), the distribution of the sample means will be approximately normal.

7.	What is the significance of confidence intervals in statistical analysis?

  Confidence intervals are crucial in statistical analysis because they provide a range of plausible values for an unknown population parameter, along with a level of confidence that this range contains the true value. They offer a more informative alternative to relying solely on point estimates and help assess the precision and reliability of statistical results.
  Here's why confidence intervals are significant:
  Quantifying Uncertainty:
  Confidence intervals acknowledge that any estimate derived from a sample is likely to be different from the true population value. They provide a range within which the true value is likely to fall, rather than just a single point estimate.
  Assessing Precision:
  A narrow confidence interval indicates a more precise estimate, suggesting that the sample data is highly representative of the population. Conversely, a wider interval suggests greater uncertainty and a less stable estimate.
  Supporting Decision-Making:
  Confidence intervals help in making informed decisions by providing a realistic range of possibilities. In A/B testing, for example, they help determine if observed differences are statistically significant or simply due to random variation.
  Interpreting Results:
  They offer a more nuanced interpretation of statistical results than just relying on p-values. A confidence interval that includes the null value (e.g., zero for a difference) suggests the result may not be statistically significant.
  Improving Validity:
  Confidence intervals can help in avoiding premature conclusions based on point estimates and promote a more cautious approach to interpreting statistical results. They ensure that the results are not just due to random chance.

8.	What is the concept of expected value in a probability distribution?

  The expected value in a probability distribution represents the average outcome of a random variable over many repetitions of an experiment. It's a weighted average, where each possible outcome is multiplied by its probability and then summed. Essentially, it's the long-term average you'd expect if you repeated the same scenario many times.





9.	Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
random_numbers = np.random.normal(mean, std_dev, 1000)

# Compute mean and standard deviation
computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

print(f"Mean: {computed_mean}, Standard Deviation: {computed_std_dev}")

# Draw a histogram
plt.hist(random_numbers, bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(computed_mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(computed_mean + computed_std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean + 1 Std Dev')
plt.axvline(computed_mean - computed_std_dev, color='green', linestyle='dashed', linewidth=1, label='Mean - 1 Std Dev')
plt.legend()
plt.show()
```

10.	You are working as a data analyst for a retail company. The company has collected daily sales
data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
 ● Write the Python code to compute the mean sales and its confidence interval.


 To apply the Central Limit Theorem (CLT) to estimate the average sales with a 95% confidence interval, you would:

Calculate the sample mean and standard deviation of the daily sales data.
Use the formula for the confidence interval:

```python
import numpy as np
from scipy import stats

daily_sales = np.array([220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
                       235, 260, 245, 250, 225, 270, 265, 255, 250, 260], dtype=float)

n = len(daily_sales)
mean_sales = np.mean(daily_sales)
s = np.std(daily_sales, ddof=1)   # sample standard deviation
df = n - 1
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df)

margin_error = t_crit * s / np.sqrt(n)
ci_lower = mean_sales - margin_error
ci_upper = mean_sales + margin_error

print(f"n = {n}")
print(f"Sample mean = {mean_sales:.2f}")
print(f"Sample sd (s) = {s:.2f}")
print(f"t-critical (df={df}) = {t_crit:.6f}")
print(f"Margin of error = {margin_error:.4f}")
print(f"95% CI (t-based): ({ci_lower:.2f}, {ci_upper:.2f})")
```
